a. Which player had the highest batting average? The lowest? Make sure to print out your results.

```python
print("The player with the Highest Batting Average was:")
display(baseball[baseball['Batting Average'] == max(baseball['Batting Average'])])
print('\n')
print("The players with the Lowest Batting Average were:")
display(baseball[baseball['Batting Average'] == min(baseball['Batting Average'])])
print('\n')
print('The player with the Lowest (Non-Zero) Batting Average was:')
display(baseball[baseball['Batting Average'] == min(baseball['Batting Average'][baseball['Batting Average'] != 0])])
print('\n Please note that this analysis (and all others below) did not require a minimum number of at-bats based on feedback from the course TAs.')
```

b. Which player had the highest # of home-runs per game?

```python
baseball['Home Runs per Game'] = baseball['Home Runs'] / baseball['Games Played']
print('The player with the Most Home Runs per Game was:')
display(baseball[baseball['Home Runs per Game'] == max(baseball['Home Runs per Game'])])
```

c. Do players with higher `Batting Average`s tend to score more `Home Runs` per game? Create a scatter plot and determine if a relationship exists (make sure to include labels!).

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(
    x = baseball['Batting Average'], 
    y = baseball['Home Runs per Game'], 
    c = "gray", 
    alpha = 0.3
    )
#sns.regplot(x = 'Batting Average', y = 'Home Runs', data = baseball)
plt.ylabel('Home Runs per Game')
plt.xlabel('Batting Average')
plt.title('Home Runs as a Function of Batting Average')
plt.show()
print('Players with higher batting averages seem to hit slightly higher numbers of home runs per game than players with lower batting averages (there is a weak-to-moderate positive correlation)')
print('The correlation between Batting Average and Home Runs is: ', np.corrcoef(baseball['Batting Average'], baseball['Home Runs per Game'])[0])
print(sns.regplot(x = 'Batting Average', y = 'Home Runs per Game', data = baseball, scatter = True))
```

3. Isolate players who play 1B and those who play outfield.

```python
baseball_filtered = baseball[baseball.POS.str.contains('OF') | baseball.POS.str.contains('1B')]
baseball_1b = baseball[baseball.POS.str.contains('1B')]
baseball_of = baseball[baseball.POS.str.contains('OF')]
print('Please note that some players play multiple positions and may therefore be included in both the "1B" and "OF" groups.')
display(baseball_filtered)
```

a. To which position does the highest home-run scoring player belong to? 

```python
print('The player with the most home runs (overall) among first basemen and outfielders plays: ', baseball_filtered.POS[baseball_filtered['Home Runs'] == max(baseball_filtered['Home Runs'])])
display(baseball_filtered[baseball_filtered['Home Runs'] == max(baseball_filtered['Home Runs'])])
print('The player with the most home runs per game among first basemen and outfielders plays: ', baseball_filtered.POS[baseball_filtered['Home Runs per Game'] == max(baseball_filtered['Home Runs per Game'])])
display(baseball_filtered[baseball_filtered['Home Runs per Game'] == max(baseball_filtered['Home Runs per Game'])])
```

b. Compare the means and medians of `Batting Average`s. Can you conclude that one group hits more successfully than the other?

```python
from scipy import stats
print('Recall that players may play more than one position and therefore may be included in multiple groups.\n')
print('The mean Batting Average for players who play 1B is: ', np.mean(baseball['Batting Average'][baseball.POS.str.contains('1B')]))
print('The median Batting Average for players who play 1B is: ', np.median(baseball['Batting Average'][baseball.POS.str.contains('1B')]))
print('The mean Batting Average for players who play OF is: ', np.mean(baseball['Batting Average'][baseball.POS.str.contains('OF')]))
print('The median Batting Average for players who play OF is: ', np.median(baseball['Batting Average'][baseball.POS.str.contains('OF')]))
print('\n') 
print('We cannot conclude that one group hits more successfully than the other, as a two-tailed independent samples t-test suggests that the mean Batting Average is not significantly different between first basemen and outfielders (p > 0.05).\n')
print(stats.ttest_ind(baseball_1b['Batting Average'], baseball_of['Batting Average'], equal_var=False))
```
